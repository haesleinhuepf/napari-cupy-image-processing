```python
from skimage.io import imread, imshow
```

```python
blobs = imread('https://samples.fiji.sc/blobs.png')
imshow(blobs)
```

```python
from napari_cupy_image_processing import gaussian_filter

blurred = gaussian_filter(blobs, sigma=3)

blurred
```

```python
from napari_cupy_image_processing import gaussian_laplace

gaussian_laplace(blobs, sigma=3)
```

```python
from napari_cupy_image_processing import median_filter

median_filter(blobs, radius=3)
```

```python
from napari_cupy_image_processing import percentile_filter

percentile_filter(blobs, percentile=90, radius=3)
```

```python
from napari_cupy_image_processing import white_tophat

white_tophat(blobs, radius=2)
```

```python
from napari_cupy_image_processing import morphological_gradient

morphological_gradient(blobs, radius=2)
```

```python
from napari_cupy_image_processing import morphological_laplace

morphological_laplace(blobs, radius=2)
```

```python
from napari_cupy_image_processing import wiener

wiener(blobs, radius=2)
```

```python
from napari_cupy_image_processing import threshold_otsu

binary = threshold_otsu(blurred)

binary
```

```python
from napari_cupy_image_processing import binary_erosion

eroded = binary_erosion(binary, iterations=5)

eroded
```

```python
from napari_cupy_image_processing import binary_dilation

dilated = binary_dilation(binary, iterations=6)

dilated
```

```python
from napari_cupy_image_processing import binary_fill_holes

filled_holes = binary_fill_holes(dilated)

filled_holes
```

```python
from napari_cupy_image_processing import label

labels = label(binary)

labels
```
