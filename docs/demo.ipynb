```python
import numpy as np
import pandas as pd
from skimage.io import imread, imshow

from napari_cupy_image_processing import threshold_otsu, gaussian_filter, label, measurements
```

```python
blobs = imread('https://samples.fiji.sc/blobs.png')
imshow(blobs)
```

```python
blurred = gaussian_filter(blobs, sigma=3)

imshow(blurred)
```

```python
binary = threshold_otsu(blurred)

imshow(binary)
```

```python
labels = label(binary)

imshow(labels, cmap="jet")
```

```python
statistics = measurements(blobs, labels)

statistics
```
